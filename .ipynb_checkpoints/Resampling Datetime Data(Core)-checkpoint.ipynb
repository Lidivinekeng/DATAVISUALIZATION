```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
df = pd.read_csv(url)
df
```

## Convert the date column to datetime dtype.

```python
pd.to_datetime(df['date'])
```

```python
#conversion 
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
```

## Set the date column as the index.

```python
df = df.set_index('date')
df
```

## Filter out only the required data:

### Keep only data from the year 2000 or later and Keep only the following features

```python
df_filtered = df.loc['2000':, ['precipitation', 'mean_temp', 'min_temp', 
                      'max_temp', 'snow_depth']]

df_filtered
```

## What month had the most precipitation between 2000 through 2010?

```python
df_filtered.plot()
```

```python
df_q1 = df_filtered['precipitation'].resample('M').sum()

df_q1
```

```python
# Find the date and corresponding value for the max precipitation
max_precipitation_date = df_q1.idxmax()

max_precipitation_date
```

```python
max_precipitation_value = df_q1.max()

max_precipitation_value 
```

```python
# Plotting Q1
fig, ax = plt.subplots(figsize=(12, 3))
df_q1.plot(ax=ax)
ax.axvline(max_precipitation_date, color='r', linestyle='--', label=f'Max Precipitation\n{max_precipitation_date.strftime("%B %Y")}: {max_precipitation_value} inches')
ax.set_title('Precipitation for 2000-2010')
ax.set_ylabel('Inches')
ax.legend()
ax.xaxis.set_major_locator(plt.MaxNLocator(6))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: pd.to_datetime(x).strftime('%Y')))
ax.xaxis.set_minor_locator(plt.MultipleLocator(3))
ax.xaxis.set_minor_formatter(plt.NullFormatter())
plt.xticks(rotation=90)
plt.show()
```

### Which year between 2000-2020 had the coolest average temperature?

```python
#Resample the  mean_temp data to a Yearly frequency using .mean() as the aggregation method.
df_q2 = df_filtered['mean_temp'].resample('Y').mean()

df_q2
```

```python
 #Find the date and value for the lowest mean temp
lowest_mean_temp_date = df_q2.idxmin()

lowest_mean_temp_date
```

```python
lowest_mean_temp_value = df_q2.min()

lowest_mean_temp_value
```

```python
# Plotting Q2
fig, ax = plt.subplots(figsize=(12, 3))
df_q2.plot(ax=ax)
ax.axvline(lowest_mean_temp_date, color='r', linestyle='--', label=f'Lowest Mean Temp\n{lowest_mean_temp_date.year}: {lowest_mean_temp_value} degrees')
ax.set_title('Average Temperature')
ax.set_ylabel('Degrees')
ax.legend()
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: pd.to_datetime(x).strftime('%Y')))
ax.xaxis.set_minor_locator(plt.MultipleLocator(1))
ax.xaxis.set_minor_formatter(plt.NullFormatter())
plt.xticks(rotation=90)
plt.show()
```
